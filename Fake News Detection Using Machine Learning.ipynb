```python
import numpy as np 
import pandas as pd 
```

```python
true = pd.read_csv('True.csv')
fake = pd.read_csv('Fake.csv')
```

```python
true['label']=1
fake['label']=0 
```

```python
news = pd.concat([fake, true],axis=0)
```

```python
news.isnull().sum()
```

```python
news.tail()
```

```python
news = news.drop(['title', 'subject', 'date'], axis=1) 
news.head()
```

```python
news = news.sample(frac=1)
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index'], axis = 1, inplace = True)
```

```python
news.head()
```

```python
import re
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub(r'http?://\S+|www\.\S+','',text) 
    text = re.sub(r'<.*?>', '',text) 
    text = re.sub(r'[^\w\s]', '',text)
    text = re.sub(r'\d','',text) 
    text = re.sub(r'\n','',text) 
    return text 
```

```python
news['text'] = news['text'].apply(wordopt) 
```

```python
news['text'] 
```

```python
x = news['text'] 
y = news['label'] 
```

```python
x 
```

```python
y 
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(x_train)
```

```python
xv_test = vectorization.transform(x_test)
```

```python
xv_test
```

```python
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
gbc.fit(xv_test, y_test)
```

```python
def output_label(n): 
    if n==0: 
        return"It is a Fake News"  
    elif n==1: 
        return "It ia a Genuine News"
```

```python
def manual_testing(news): 
    testing_news = {"text": [news]} 
    new_def_test = pd.DataFrame(testing_news) 
    new_def_test["text"] = new_def_test["text"].apply(wordopt) 
    new_x_test = new_def_test["text"] 
    new_xv_test = vectorization.transform(new_x_test) 
    pred_lr = LR.predict(new_xv_test) 
    pred_dtc = DTC.predict(new_xv_test) 
    pred_gbc = gbc.predict(new_xv_test) 
    pred_rfc = rfc.predict(new_xv_test) 
    return "\n\nLR Prediction : {} \nDTC Prediction : {} \nGBC Prediction: {} \n RFC Prediction: {}". format(
        output_label(pred_lr[0]), output_label(pred_dtc[0]), output_label(pred_gbc[0]), output_label(pred_rfc[0])
    ) 
```

```python
news_article = """The media has been talking all day about Trump and the Republican Party scam of a tax bill; as well as the sheer obsequiousness of True"""
```

```python
manual_testing(news_article)
```
